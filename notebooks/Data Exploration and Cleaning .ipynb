# data_exploration 

```python
import pandas as pd
import json

# Load the datasets
csv_data = pd.read_csv('data/raw data /Accidents-2019.csv')

with open('data/raw data /CalderdaleCollisions2020.json', 'r') as file:
    json_data = pd.json_normalize(json.load(file))

# Preview the datasets
print("CSV Data Preview:")
display(csv_data.head())

print("\nJSON Data Preview:")
display(json_data.head())
```

```python
print(csv_data.info())
print(json_data.info())
```

```python
print("CSV Missing Data Summary:")
print(csv_data.isnull().sum())

print("\nJSON Missing Data Summary:")
print(json_data.isnull().sum())

```

```python
print("Unique Road Surface Conditions in CSV:")
print(csv_data['Road Surface'].unique())

print("\nUnique Road Surface Conditions in JSON:")
print(json_data['Road Surface'].unique())
```

# Data Exploration Summary

## CSV Dataset
1. **Shape**: 271 rows and 45 columns.
2. **Missing Values**:
   - Significant missing data in columns related to additional vehicles and casualties:
     - `Type of Vehicle 2`: 80 missing values.
     - `Type of Vehicle 3`: 232 missing values.
     - `Type of Vehicle 4-7`: Majority of rows are missing.
   - First vehicle and casualty details (`Type of Vehicle 1`, `Casualty Severity 1`, etc.) have no missing data.
3. **Unique Road Surface Conditions**:
   - Categories: `['Dry', 'Wet / Damp', 'Frost / Ice', 'Flood (surface water over 3cm deep)', 'Snow']`.


## JSON Dataset
1. **Shape**: 202 rows and 40 columns.
2. **Missing Values**:
   - Only the `Reference Number` column has 2 missing values.
   - All other columns have no missing data.
3. **Unique Road Surface Conditions**:
   - Categories:
     - Clean categories: `['Wet / damp', 'Dry', 'Frost / ice', 'Snow']`.
     - Duplicated formatting issues: `['Wet  / damp', 'Wet /  damp']`.


## Initial Observations
1. The JSON dataset is more complete but has formatting inconsistencies in `Road Surface`.
2. The CSV dataset has more missing values, especially for secondary vehicles and casualties.
3. Both datasets have some overlap in the structure (e.g., casualty and vehicle details) but may need cleaning for consistent formatting.


# data cleaning 
## Step 1: Planning the Cleaning Tasks

```python
# Check for missing values
print("Missing Values in CSV Dataset:")
print(csv_data.isnull().sum())

print("\nMissing Values in JSON Dataset:")
print(json_data.isnull().sum())

# Check for duplicates
print("\nDuplicate Rows in CSV Dataset:", csv_data.duplicated().sum())
print("Duplicate Rows in JSON Dataset:", json_data.duplicated().sum())

# Check unique values for key columns
print("\nUnique Road Surface Conditions in CSV:")
print(csv_data['Road Surface'].unique())

print("\nUnique Road Surface Conditions in JSON:")
print(json_data['Road Surface'].unique())
```

## Step 2: Handle Missing Values


```python
# Handle missing values in CSV dataset
# Drop columns with excessive missing data 
csv_data_cleaned = csv_data.drop(columns=[
    'Type of Vehicle 4', 'Casualty Class 4', 'Casualty Severity 4', 'Sex of Casualty 4', 'Age of Casualty 4',
    'Type of Vehicle 5', 'Casualty Class 5', 'Casualty Severity 5', 'Sex of Casualty 5', 'Age of Casualty 5',
    'Type of Vehicle 6', 'Casualty Class 6', 'Casualty Severity 6', 'Sex of Casualty 6', 'Age of Casualty 6',
    'Type of Vehicle 7', 'Casualty Class 7', 'Casualty Severity 7', 'Sex of Casualty 7', 'Age of Casualty 7'
])

# Replace missing values in Type of Vehicle 2 and 3 with 'Unknown'
csv_data_cleaned['Type of Vehicle 2'].fillna('Unknown', inplace=True)
csv_data_cleaned['Type of Vehicle 3'].fillna('Unknown', inplace=True)

# Handle missing values in JSON dataset
# Drop rows where Reference Number is missing
json_data_cleaned = json_data.dropna(subset=['Reference Number'])

# Preview cleaned datasets
print("CSV Data after handling missing values:")
print(csv_data_cleaned.info())

print("\nJSON Data after handling missing values:")
print(json_data_cleaned.info())

```

## Step 3: Standardize Formatting

```python
# Standardize Road Surface in JSON dataset
json_data_cleaned['Road Surface'] = json_data_cleaned['Road Surface'].str.strip().replace({
    'Wet  / damp': 'Wet / Damp',
    'Wet /  damp': 'Wet / Damp',
    'Wet / damp': 'Wet / Damp'
})

# Preview unique values in Road Surface after standardization
print("Unique Road Surface Conditions in JSON (after cleaning):")
print(json_data_cleaned['Road Surface'].unique())
```

## step 4 :Validate the Cleaned Data

```python
# Check for duplicates in the cleaned datasets
print("Duplicate Rows in CSV Dataset (after cleaning):", csv_data_cleaned.duplicated().sum())
print("Duplicate Rows in JSON Dataset (after cleaning):", json_data_cleaned.duplicated().sum())

# Confirm data types in both datasets
print("\nCSV Dataset Data Types:")
print(csv_data_cleaned.dtypes)

print("\nJSON Dataset Data Types:")
print(json_data_cleaned.dtypes)
```

## step 5 : Save the Cleaned Datasets

```python
# Save the cleaned CSV dataset
csv_data_cleaned.to_csv('data/cleaned data /Accidents-2019-cleaned.csv', index=False)

# Save the cleaned JSON dataset
json_data_cleaned.to_json('data/cleaned data /CalderdaleCollisions2020-cleaned.json', orient='records')

print("Cleaned datasets saved successfully!")
```
